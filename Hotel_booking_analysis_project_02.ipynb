### Handling Outliers

```python
# Getting a list of categorical variables
categorical_var=[i for i in df.columns if df[i].dtype=='O']
print(f'There are {len(categorical_var)} categorical variables present in the dataset\n')
print(categorical_var)
```

```python
# Getting a list of numerical variables
numerical_var=[i for i in df.columns if df[i].dtype!='O']
print(f'There are {len(numerical_var)} numerical variables present in the dataset\n')
print(numerical_var)
```

```python
# Getting a list of discrete variables from list of numerical variables
# Here we consider variables having less than 150 categories as discrete variable
discrete_var=[]
for i in numerical_var:
  if len(df[i].value_counts())<=150:
    discrete_var.append(i)
    print(i,':-',df[i].unique())
    print('--'*50,'\n')
  else:
    continue
print(f'There are {len(discrete_var)} discrete variables in the dataset\n')
```

```python
# Getting list of continuous  variables from the list of numerical variables
continuous_var=[i for i in numerical_var if i not in discrete_var]
print(f'There are {len(continuous_var)} continuous varibles present in dataset')
print('--'*45,'\n')
print(continuous_var)
```

```python
# Checking outliers for the continuous variables
plt.figure(figsize=(20,4))
plt.subplot(1,3,1)
sns.boxplot(data=df,y=df['lead_time'])
plt.title('lead-time')
plt.show()
print('_'*150,'\n')

plt.figure(figsize=(20,4))
plt.subplot(1,3,2)
sns.boxplot(data=df,y=df['agent'])
plt.title('Agent')
plt.show()
print('_'*150,'\n')

plt.figure(figsize=(20,4))
plt.subplot(1,3,3)
sns.boxplot(data=df,y=df['adr'])
plt.title('Adr')
plt.show()
```

Here we found outliers in lead_time and adr and not in agent. So perform some operation on lead_time and adr variables to remove outliers.

```python
# Removing outliers from lead_time and adr variables
# Outlier columns
outliers_columns=['lead_time','adr']

# Copy dataset to new dataset
new_df=df.copy()

for i in outliers_columns:
    #Findng IQR
    Q1=new_df[i].quantile(0.25)
    Q3=new_df[i].quantile(0.75)
    IQR=Q3-Q1

    # Defining lower and upper limit
    lower_limit =new_df[i].quantile(0.25)-1.5*IQR
    upper_limit =new_df[i].quantile(0.75)+1.5*IQR

    # Applying lower and upper limit to each variables
    new_df.loc[(new_df[i] > upper_limit),i] = upper_limit
    new_df.loc[(new_df[i] < lower_limit),i] = lower_limit
```

```python
# Checking for outliers for both the columns
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
sns.boxplot(data=new_df,y=new_df['lead_time'])
plt.title('lead-time')
plt.show()
print('_'*150,'\n')
plt.figure(figsize=(20,5))
plt.subplot(1,2,2)
sns.boxplot(data=new_df,y=new_df['adr'])
plt.title('Adr')
plt.show()
```

We have successfully removed outliers from both the  columns (lead_time and adr).

### **Data Wrangling**

```python
# Here we perform some data wrangling operations to make dataset ready for analysis.
# Converting columns to appropriate datatypes.
#  Change datatype of variables children and agent to correct format from float64 to int64
new_df[['children','agent']]=new_df[['children','agent']].astype(int)

# Change datatype of variable reservation_status_date to correct format from object to datetime64
new_df['reservation_status_date']=pd.to_datetime(new_df['reservation_status_date'],format='%Y-%m-%d')
```

```python
# Check new updated datatype
new_df[['children','agent','reservation_status_date']].info()
```

Data types are successfully converted from float64 to int64 for children and agent variable and from object to datetime64 for reservation_status_date variable.




```python
# Adding important columns.

# Creating "total_children" variable by adding children and babies variables
new_df['total_children']=new_df['children']+new_df['babies']

# Creating total_people variable by adding adults,children and babies variables
new_df['total_people']=new_df['adults']+new_df['children']+new_df['babies']

# Creating a variable "total_stay" by adding stays_in_week_nights and stays_in_weekend_nights variables
new_df['total_stay']=new_df['stays_in_weekend_nights']+new_df['stays_in_week_nights']

# Creating 'reserved_room_assigned' variable which describe same room assigned or not by comparing two variables 'reserved_room_type' and 'assigned_room_type'
new_df['reserved_room_assigned']=np.where(new_df['reserved_room_type']==new_df['assigned_room_type'],'Yes','No')

# Creating 'guest_category' varible from variable 'total_people'
new_df['guest_category']=np.where(new_df['total_people']==1, 'single',
                                 np.where(new_df['total_people']==2, 'couple', 'family'))

```

In above section, we created some new columns by adding or comparing existing columns to make better analysis. We also use numpy library to perform some comapring operation and assign value to new column on the basic of condition.
